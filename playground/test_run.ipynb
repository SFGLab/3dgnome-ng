```python
from gnome3d.simulate import run_region

structures = run_region(
    config_path="../data/GM12878/config.ini",
    data_dir="../data/GM12878",
    region="chr1:18288319-20307135",
    n_structures=1
)
```

```python
structure = structures[0]
print(len(structure))
for bead in structure:
    print(f"start={bead.start}, end={bead.end}, span={bead.span}, kind={bead.kind}, x={bead.x:.3f}, y={bead.y:.3f}, z={bead.z:.3f}")
```
